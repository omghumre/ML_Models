## Decision tree

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
df.shape
```

```python
df.head()
```

```python

x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
xtrain, xtest, ytrain, ytest = train_test_split(x,y, test_size=0.3, random_state=1)
```

```python
def mymodel(model):
    model.fit(xtrain,ytrain)
    ypred=model.predict(xtest)
    print(classification_report(ytest,ypred))
    return model
```

```python
dt=DecisionTreeClassifier()
```

```python
mymodel(dt)
```

```python
dt.feature_importances_
```

## Hyperparameter tuning

```python
dt1=DecisionTreeClassifier(max_depth=10)
mymodel(dt1)
```

```python
for i in range(1,50):
    dt2=DecisionTreeClassifier(max_depth=i)
    dt2.fit(xtrain,ytrain)
    ypred=dt2.predict(xtest)
    print(f"{i}: {accuracy_score(ytest,ypred)}")
    print()
```

```python
dt3=DecisionTreeClassifier(max_depth=20)
mymodel(dt3)
```

```python
for i in range(1,75):
    dt2=DecisionTreeClassifier(min_samples_leaf=i)
    dt2.fit(xtrain,ytrain)
    ypred=dt2.predict(xtest)
    print(f"{i}: {accuracy_score(ytest,ypred)}")
```

```python
from sklearn import tree
fig = plt.figure(figsize=(25,20))
t= tree.plot_tree(dt3,filled=True,feature_names=x.columns)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
from sklearn.model_selection import cross_val_score
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(accuracy_score(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
accuracy_rate = []

for i in range(1,40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred = knn.predict(X_test)
    score = accuracy_score(y_test,pred)
    accuracy_rate.append(score)
```

```python
accuracy_rate
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),accuracy_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Accuracy Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Accuracy Rate')
plt.show()
```

```python
# FIRST A QUICK COMPARISON TO OUR ORIGINAL K=2
knn = KNeighborsClassifier(n_neighbors=2)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=1')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(accuracy_score(y_test,pred))
```

```python
# NOW WITH K=5
knn = KNeighborsClassifier(n_neighbors=21)
knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=21')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(accuracy_score(y_test,pred))
```
